*   모듈에는 변수 선언, 함수 정의, 클래스 정의 등이 포함됨.



```
import matplotlib.pyplot as plt
# matplotlib 패키지의 pyplot 모듈을 plt라는 별명으로 import, 사용.

import seaborn as sns
```

```
# seaborn 패키지의 샘플 데이터 프레임 tips를 로드
tips = sns.load_dataset('tips')
tips.iloc[:5]  # tips.head()
```

**matplotlib.pyplot 모듈을 사용한 시각화(visualization)**

```
# 1) figure 객체 생성.
fig = plt.figure()

# 2) figure 객체에 subplot을 생성해서 추가.
axis = fig.add_subplot(1, 1, 1)

# 3) subplot에 그래프를 그림.
axis.hist(x=tips['total_bill'])

# 4) 그래프 옵션 설정
axis.set_title('Histogram of Total Bill')
axis.set_xlabel('Total Bill')
axis.set_ylabel('# of bills')

# 5) 그래프 보여주기
plt.show()
```

```
# 1개의 figure에 2개 (이상)의 subplot를 그림.
# 1) figure 생성.
fig = plt.figure()

# 2) subplot을 생성, 추가: add_subplot(nrow, ncol, index), index = 1, 2, 3, ...
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# 3) subplot에 그래프 그림.
ax1.hist(x=tips['total_bill'])
ax2.scatter(x=tips['total_bill'], y=tips['tip'])

# 4) 옵션 설정
ax1.set_title('Histogram of Total Bill')
ax2.set_title('Scatter Plot(Total Bill vs Tip)')

# 5) 그래프 보여주기.
plt.show()
```

```
# 1) figure 생성.
fig = plt.figure()

# 2) subplot 생성, 추가.
# add_subplot(nrow, ncol, index)
# add_subplot(3-digit number): 100-nrow, 10-ncol,1-index
ax1 = fig.add_subplot(211)  # fig.add_subplot(2, 1, 1)
ax2 = fig.add_subplot(212)  # fig.add_subplot(2, 1, 2)

# 3) 그래프 그림.
ax1.hist(x=tips['total_bill'])
ax2.scatter(x=tips['total_bill'], y=tips['tip'])

# 4) 옵션 설정
# 5) 그래프 보여주기.
plt.show()
```

```
# matplotlib.pyplot.subplots(nrows, ncols): 
# figure 생성, subplot 생성 및 추가 => (figure, array[subplot1, suplot2, ...]) 튜플을 리턴.
fig, axes = plt.subplots(nrows=2, ncols=1)

# 각 subplot에 그래프를 그림.
axes[0].hist(x=tips['total_bill'])
axes[1].scatter(x=tips['total_bill'], y=tips['tip'])

# 옵션 설정
axes[0].set_title('Histogram')
axes[1].set_title('Scatter Plot')

# 그래프 보여주기
plt.show()
```

```
# 1개의 figure에 subplot이 1개만 있는 경우에는,
# matplotlib.pyplot 모듈의 top-level 함수(hist, scatter, boxplot, ...)들만 사용해도 됨.
plt.hist(x=tips['total_bill'])
plt.show()
```

**matplotlib.pyplot의 그래프 종류들**

```
import numpy as np

data = np.random.rand(10)
print(data)
```

```
# 선 그래프(plot)
plt.plot(data)
plt.show()
```

```
plt.plot(data, 'bo--')  # b(blue: color), o(marker), --(linestyle)
plt.show()
```

```
plt.plot(data, color='green', marker='s', linestyle='dotted')
plt.show()
```

```
x_pts = np.arange(0, 100, 10)
y_pts = np.random.rand(10)

plt.plot(x_pts, y_pts, marker='+')
plt.show()
```

```
# 1개 subplot에 여러개의 선 그래프(plot) 그리기.

np.random.seed(1)

y1 = np.random.rand(10)
y2 = np.random.rand(10)
y3 = np.random.rand(10)

plt.plot(y1, color='red', linestyle='solid', label='One')
plt.plot(y2, color='green', linestyle='dotted', label='Two')
plt.plot(y3, color='blue', linestyle='dashed', label='Three')

plt.legend()  # 범례(legend) 보여주기 - plot 함수에서 설정한 label을 보여줌.

plt.show()
```

```
# pandas.Series, pandas.DataFrame의 describe() 메서드: 기술 통계량
# R의 summary() 함수와 비슷.
tips['tip'].describe()
```

```
# boxplot: 기술 통계량 요약(최솟값, 25%, 중위값, 75%, 최댓값)
plt.boxplot(x=tips['tip'], labels=['tip'])
plt.show()
```

```
tips.describe()  # df.describe(): 숫자 타입 컬럼들의 기술 통계량 요약 정보.
```

```
# 1개의 subplot 안에 total_bill, tip 컬럼의 box plot을 그림.
plt.boxplot(x=[tips['total_bill'], tips['tip']],
            labels=['total_bill', 'tip'])
plt.show()
```

```
# 성별 tip의 기술 통계량 요약
# tips['sex'].value_counts()  # 성별: Male, Female
male_tip = tips[tips['sex'] == 'Male']['tip']
print(male_tip.describe())

female_tip = tips[tips['sex'] == 'Female']['tip']
print(female_tip.describe())

# 성별 tip의 box plot
plt.boxplot(x=[male_tip, female_tip],
            labels=['Male', 'Female'])
plt.show()
```

```
# 요일별 tip의 기술 통계량 요약
# tips['day'].value_counts()
thur_tip = tips[tips['day'] == 'Thur']['tip']
fri_tip = tips[tips['day'] == 'Fri']['tip']
sat_tip = tips[tips['day'] == 'Sat']['tip']
sun_tip = tips[tips['day'] == 'Sun']['tip']

# 요일별 tip의 box plot
plt.boxplot(x=[thur_tip, fri_tip, sat_tip, sun_tip],
            labels=['Thur', 'Fri', 'Sat', 'Sun'])
plt.show()
```

```
days = tips['day'].value_counts().index
for day in days:
    print(f"===== {day}=====\n{tips[tips['day'] == day]['tip'].describe()}")
```

```
# 히스토그램(histogram):
# 연속형 변수를 (일정한) 구간으로 나눠서, 그 구간 안에 포함된 데이터의 개수를 막대로 표현한 그래프.
# bins: 막대 개수(bins 기본값 10)
# density: 밀도 추정 여부(density=False: 데이터 개수, density=True: 밀도)
plt.hist(x=tips['total_bill'], bins=10, density=True)
plt.show()
```

```
# 산점도 그래프(scatter plot): 두 변수 간의 상관 관계(y ~ x)를 파악.
# tip ~ total_tip 관계
plt.scatter(x=tips['total_bill'], y=tips['tip'],
            s=50,         # marker size
            c='rebeccapurple',  # marker color: 색깔 이름, #RRGGBB(16진수)
            alpha=0.3     # 불투명도: 0(투명, transparent) ~ 1(불투명, opaque)
            )
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```
# plot() 함수: 선 그래프(line graph), 산점도 그래프(linestyle을 제거, marker 지정)
plt.plot(tips['total_bill'], tips['tip'],
         marker='o', linestyle='', color='green', markersize=5)
plt.show()
```

```
# 막대 그래프(bar plot): 범주형 변수(categorical variable)의 빈도수를 막대로 표현
gender_count = tips['sex'].value_counts()
print(gender_count)

plt.bar(gender_count.index, gender_count.values)
plt.show()
```
