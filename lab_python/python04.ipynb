list: 여러개의 값들을 하나의 변수에 저장하기 위한 데이터 타입.<br>
원소(element): 리스트에 저장된 값.<br>
인덱스(index): 리스트에 값이 저장된 위치(번호). 인덱스는 0부터 시작.<br>
리스트는 원소들을 변경(추가/삭제)이 가능함. - tuple과의 차이점!

```
numbers = [1, 2, 3, 4, 5]
print(numbers, type(numbers))
```

```
# list_name[index]: 해당 index 위치에 있는 원소를 선택
print(numbers[0])  # 리스트에서 인덱스는 0부터 시작.
print(numbers[1])
print(numbers[4])  # 리스트에서 마지막 인덱스는 (원소의 개수 - 1).
```

```
# list_name[start:end]: start에서부터 (end - 1)까지의 원소들을 선택. start <= index < end.
print(numbers[0:3])  # 0 <= index < 3
print(numbers[2:5])  # 2 <= index < 5
```

```
# 특정 인덱스에 있는 원소의 값을 변경: list[index] = value
numbers[0] = 123
print(numbers)
```

```
# list에 원소를 추가: append(), extend() 메서드(함수)
numbers.append(6)
print(numbers)
```

```
numbers.extend([7])  # extend(list/tuple)
print(numbers)
```

```
numbers.append([8, 9, 10])
print(numbers)
#> [8, 9, 10] 리스트가 원소로 추가됨!
```

```
numbers.extend([11, 12, 13])
print(numbers)
#> 리스트의 각 원소들(11, 12, 13)이 차례로 추가됨!
```

```
# 리스트의 원소 삭제: remove() 메서드
# list.remve(value)
numbers = [10, 20, 30, 10, 50]
print(numbers)

numbers.remove(10)
print(numbers)

# numbers.remove(100)  # 리스트에 없는 값을 삭제하려고 할 때 에러 발생.
```

```
# 리스트는 여러가지 타입의 값들을 함께 저장할 수 있음.
person = ['오쌤', 16, 123.456, True]
print(person, type(person))
print(person[0], type(person[0]))
print(person[1], type(person[1]))
```

```
student = [1, 'abc']
print(student[0])  #> 1
print(student[1])  #> abc

# list decomposition
stu_no, stu_name = student
print(stu_no)  #> 1
print(stu_name)  #> abc
```

```
# 2차원 리스트: 1차원 리스트를 원소로 갖는 리스트.
matrix = [
          [1, 2, 3],
          [4, 5, 6],
          [7, 8, 9]
]
print(matrix)
print(matrix[0], type(matrix[0]))
print(matrix[0][1])  # 2d_list[row][column]
print(matrix[0:2])  # 0 <= row < 2
```

```
numbers = [1, 2, 3, 4, 5]
print(numbers[-2])
print(numbers[-2:])
```

튜플(tuple): 원소들을 변경할 수 없는 리스트. <br>
인덱스를 사용해서 저장된 원소를 확인(선택).


```
numbers = (1, 2, 3, 4, 5)
print(numbers, type(numbers))
print(numbers[0])  #> 1
print(numbers[0:3])  # 0 <= index < 3
print(numbers[-2:])  #> (4, 5)
```

```
# numbers[0] = 123
# tuple에 저장된 값은 변경할 수 없음!
# tuple은 append(), extend(), remove()와 같은 메서드(기능)을 제공하지 않음.
```

```
# tuple decomposition
one, two, three = (1, 2, 3)
print(one)
print(two)
print(three)
```
