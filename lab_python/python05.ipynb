```
# 문자열(str): 문자들의 리스트
# 인덱스를 사용할 수 있음.
# 인덱스를 이용한 slicing 기능도 제공.

msg = 'Hello, Python!'
print(msg)
print(msg[0])
print(msg[5])
print(msg[7:13])
print(msg[-3:])
```

```
msg.capitalize()  # 문자열의 첫글자만 대문자로, 나머지는 소문자로 변환
```

```
msg.upper()  # 모든 문자를 대문자로 변환
```

```
msg.lower()  # 모든 문자를 소문자로 변환
```

```
observation = '1,Hello,100,Male'  # 문자열
observation.split(',')
```

```
# 문자열은 +, * 연산이 가능함.
s1 = 'hello'
s2 = 'python'

print(s1 + s2)  
# 문자열 + 문자열: concatenate. 두 개의 문자열을 이어붙여주는 기능.
# 문자열 + 숫자, 숫자 + 문자열: 에러 발생!
# print(1 + s1)  # 에러발생

print(s1 * 3)
# 문자열 * 숫자: 문자열을 숫자만큼 반복.
```
