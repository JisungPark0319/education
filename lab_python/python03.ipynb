**Python 데이터 타입**

*   숫자 타입: int(정수), float(실수), complex(복소수)
*   논리 타입(True/False): bool
*   문자열 타입: str
*   시퀀스(sequence) 타입: list, tuple
*   매핑(mapping) 타입: dict
*   집합 타입: set
*   None: 값이 없음을 나타내는 데이터 타입.



```
n = 100
print(type(n))
```

```
x = 3.141592
print(type(x))
```

```
# 복소수 = 실수부 + 허수부j
c = 1j
print(type(c))
print(c ** 2)  # 허수: 제곱을 했을 때 음수가 되는 수
```

```
# Python에서 문자열은 큰따옴표("") 또는 작은따옴표('')를 사용.
s1 = "Hello"
s2 = 'Hello'
print(type(s1), type(s2))
```

```
message = 'I\'m a boy.'
print(message)

message2 = "I'm a girl."
print(message2)

message3 = '오쌤이 "안녕하세요!"라고 말했습니다.'
print(message3)
```

```
result = 2 < 5
print(result, type(result))
```

*   비교 연산자: ==, !=, >, >=, <, <=
*   논리 연산자: and, or, not

```
result = (2 < 5) and (2 < 0)  # True and False
print(result)
```

```
names = ['김상경', '김수연', '김수현']  # list
print(names, type(names))
```

```
names2 = ('김원준', '김정욱', '김태현')  # tuple
print(names2, type(names2))
```

```
students = {1: '김상경', 2: '김수연', 3: '김수현'}  # dict
print(students, type(students))
# key: value 쌍으로 mapping해서 저장하는 데이터 타입
```
